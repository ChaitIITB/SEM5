```python
import matplotlib.pyplot as plt
import numpy as np

def plotline(w, theta, title='Data Plot', xlabel='x2', ylabel='x1'):
    x = np.arange(-10, 10, 0.1)
    y = [((w[0] * xi + theta)/w[1]) * -1 for xi in x]
    plt.plot(x, y, label='Line')
    
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('center')

    # Eliminate upper and right axes
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    # Show ticks in the left and lower axes only
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    plt.title(title)
    plt.show()
```

```python
plotline([1, -1], 0, title='Example Line Plot', xlabel='X-axis', ylabel='Y-axis')
# This script plots a line based on the provided weights and theta values.
```
